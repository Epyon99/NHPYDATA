```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Cargar los datos del archivo CSV en un DataFrame de Pandas
df = pd.read_csv('datapersonas.csv')

# Mostrar los datos
# print(df)

# Parámetros de la distribución normal (usamos la columna 'Age')
mean_age = df['Age'].mean()
std_age = df['Age'].std()

# Generar 1000 muestras aleatorias con distribución normal
normal_age = np.random.normal(loc=mean_age, scale=std_age, size=1000)
print ('Preplot')
print (type(normal_age))
print (len(normal_age))
print ('---fin----preplot')
# Graficar la distribución normal
plt.hist(normal_age, bins=30, density=True, alpha=0.6, color='g')
plt.title('Distribución Normal de las Edades')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Cargar los datos del archivo CSV en un DataFrame de Pandas
df = pd.read_csv('datapersonas.csv')

# Mostrar los datos
print(df)

# Parámetros de la distribución binomial
n = 100  # número de ensayos
p = len(df[df['Weight'] > 70]) / len(df)  # probabilidad de éxito (más de 70 kg)

# Generar 1000 muestras aleatorias con distribución binomial
binomial_weight = np.random.binomial(n, p, 1000)
print (binomial_weight)
# Graficar la distribución binomial
plt.hist(binomial_weight, bins=30, density=True, alpha=0.6, color='b')
plt.title('Distribución Binomial del Peso > 70kg')
plt.xlabel('Número de éxitos (peso > 70kg)')
plt.ylabel('Frecuencia')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Cargar los datos del archivo CSV en un DataFrame de Pandas
df = pd.read_csv('datapersonas.csv')

# Mostrar los datos
print(df)

# Parámetro de la distribución Poisson (usamos la cantidad de personas con peso > 70)
lambda_poisson = len(df[df['Weight'] > 70])  # tasa de eventos (personas > 70 kg)

# Generar 1000 muestras aleatorias con distribución Poisson
poisson_weight = np.random.poisson(lambda_poisson, 1000)

# Graficar la distribución Poisson
plt.hist(poisson_weight, bins=30, density=True, alpha=0.6, color='r')
plt.title('Distribución Poisson del Peso > 70kg')
plt.xlabel('Número de eventos (personas > 70kg)')
plt.ylabel('Frecuencia')
plt.show()

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Cargar los datos del archivo CSV en un DataFrame de Pandas
df = pd.read_csv('datapersonas.csv')

# Mostrar los datos
print(df)

# Parámetro de la distribución exponencial (media de la altura)
mean_height = df['Height'].mean()

# Generar 1000 muestras aleatorias con distribución exponencial
exponential_height = np.random.exponential(scale=mean_height, size=1000)

# Graficar la distribución exponencial
plt.hist(exponential_height, bins=30, density=True, alpha=0.6, color='y')
plt.title('Distribución Exponencial de la Altura')
plt.xlabel('Altura')
plt.ylabel('Frecuencia')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Cargar los datos del archivo CSV en un DataFrame de Pandas
df = pd.read_csv('datapersonas.csv')

# Mostrar los datos
print(df)

# Parámetros de la distribución uniforme (edad mínima y máxima)
min_age = df['Age'].min()
max_age = df['Age'].max()

# Generar 1000 muestras aleatorias con distribución uniforme
uniform_age = np.random.uniform(low=min_age, high=max_age, size=1000)

# Graficar la distribución uniforme
plt.hist(uniform_age, bins=30, density=True, alpha=0.6, color='c')
plt.title('Distribución Uniforme de las Edades')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show()
```
